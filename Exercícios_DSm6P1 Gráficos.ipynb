```python
import pandas as pd
import matplotlib.pyplot as plt

# Gráfico de barras para APGAR1
plt.figure(figsize=(8, 4)) 
df_sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribuição do APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')
plt.show()

# Gráfico de barras para APGAR5
plt.figure(figsize=(8, 4))
df_sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('Distribuição do APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')
plt.show()

# Gráfico de barras para SEXO
plt.figure(figsize=(8, 5))
df_sinasc['SEXO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribuição do SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')
plt.show()

# Gráfico de barras para GESTACAO
plt.figure(figsize=(8, 4))
df_sinasc['GESTACAO'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Distribuição do GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')
plt.show()

# Gráfico de barras para ESTCIVMAE
plt.figure(figsize=(8, 4))
df_sinasc['ESTCIVMAE'].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('Distribuição do ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calcular a média de APGAR5 para cada valor único em CONSULTAS
media_apgar5_consultas = df_sinasc.groupby('CONSULTAS')['APGAR5'].mean()

# Criar o gráfico de barras
plt.figure(figsize=(8,4))
media_apgar5_consultas.plot(kind='bar', color='skyblue')

# Definir o título a rótulos dos eixos
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')

# Mudar a rotação dos rótulos dos valores de CONSULTAS para 0 graus
plt.xticks(rotation=0)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular a média de APGAR5 para cada valor único em CONSULTAS
media_apgar5_consultas = df_sinasc.groupby('CONSULTAS')['APGAR5'].mean()

# Escolha uma paleta de cores verde do Seaborn
cores_verde = sns.color_palette('Greens')

# Criar o gráfico de barras
plt.figure(figsize=(8,4))
media_apgar5_consultas.plot(kind='bar', palette=cores_verde)

# Definir o título a rótulos dos eixos
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')

# Mudar a rotação dos rótulos dos valores de CONSULTAS para 0 graus
plt.xticks(rotation=0)

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import seaborn as sns

# Defina um layout de subplots para o painel com 2 linhas e 2 colunas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14,10))

# Gráfico de frequência de SEXO
sns.countplot(data=df_sinasc, x='SEXO', ax=axes[0,0]) # Cria um gráfico de contagem de SEXO no primeiro subplot
axes[0, 0].set_title('Frequência de SEXO') # Define o título para o primeiro gráfico

# Gráfico de frequência de GESTACAO
sns.countplot(data=df_sinasc, x='GESTACAO', ax=axes[0, 1]) # Cria um gráfico de contagem de GESTACAO no segundo subplot
axes[0, 1].set_title('Frequência de GESTACAO') # Define o título para o segundo gráfico
plt.xticks(rotation=0) # Não funcionou rotacionar o periodo de semanas, nao sei porque


# Gráfico de frequência de ESTCIVMAE
sns.countplot(data=df_sinasc, x='ESTCIVMAE', ax=axes[1, 0]) # Cria um gráfico de contagem de ESTCIVMAE no terceiro subplot
axes[1, 0].set_title('Frequência de ESTCIVMAE')  # Define o título para o terceiro gráfico
axes[1, 0].tick_params(axis='x', rotation=45)

# Gráfico de frequência de CONSULTAS
sns.countplot(data=df_sinasc, x='CONSULTAS', ax=axes[1, 1]) # Cria um gráfico de contagem de CONSULTAS no quarto subplot
axes[1, 1].set_title('Frequência de CONSULTAS') # Define o título para o quarto gráfico
axes[1, 1].tick_params(axis='x', rotation=45)

# Ajusta o espaçamento entre os gráficos para evitar sobreposições
plt.tight_layout()

plt.show()
```
