```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o arquivo CSV
df_sinasc = pd.read_csv('sinasc_RO_2019.csv')

df_sinasc
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Crie o boxplot
plt.figure(figsize=(8, 6)) # Tamanho da figura
sns.boxplot(data=df_sinasc, y='IDADEMAE', color='Purple') # Cria o boxplot
plt.title('Boxplot da Idade da Mãe') # Adiciona o titulo do grafico
plt.ylabel('Idade da Mãe') # titulo do eixo y (vertical)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6)) # tamanho da figura
sns.boxplot(data=df_sinasc, y='PESO', color='skyblue') # Cria o boxplot e define a cor
plt.title('Boxplot do Peso do Bebê') # Título do grafico
plt.ylabel('Peso do Bebê (em gramas)') # Titulo do eixo y (vertical)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Extraia a coluna 'IDADEMAE' do df_sinasc
idade_maes = df_sinasc['IDADEMAE']

# Cria o histograma
plt.hist(idade_maes, bins=20, color='skyblue', edgecolor='black')

# Adiciona os titulos
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.title('Histograma da Idade das Mães no SINASC')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
peso_do_bebe = df_sinasc['PESO']

# Crie um histograma
plt.figure(figsize=(10, 6))
plt.hist(peso_do_bebe, bins=20, color='skyblue')

# Adicione rotulos e titulo ao grafico
plt.xlabel('Peso do bebê (em gramas)')
plt.ylabel('Frequência')
plt.title('Histograma do Peso do bebê')

plt.show()
```
